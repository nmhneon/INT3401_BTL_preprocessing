```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
from itertools import product
plt.rcParams['figure.figsize'] = [3,4]
```

```python
image0=cv2.imread('2.jpg')
height,width,_=image0.shape
newH,newW=int(height*1500/width),1500
image0=cv2.resize(image0,(newW,newH))
image=cv2.cvtColor(image0,cv2.COLOR_BGR2GRAY)
plt.imshow(image,cmap='gray')
```

```python
dilate=cv2.dilate(image,np.ones((7,7), np.uint8))
plt.imshow(dilate,cmap='gray')
```

```python
bg_img = cv2.medianBlur(dilate, 21)
plt.imshow(bg_img,cmap='gray')
```

```python
diff_img = 255 - cv2.absdiff(image, bg_img)
plt.imshow(diff_img,cmap='gray')
cv2.imwrite('out1.jpg',diff_img)
```

```python
ret,img1=cv2.threshold(diff_img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.imshow(img1,cmap='gray')
```

```python
gradX = cv2.Sobel(img1,cv2.CV_32F, dx=1, dy=0, ksize=-1)
gradY = cv2.Sobel(img1,cv2.CV_32F, dx=0, dy=1, ksize=-1)
image3=cv2.subtract(gradX,gradY)
image3=cv2.convertScaleAbs(image3)
plt.imshow(image3,cmap='gray')
cv2.imwrite('out12.jpg',image3)
```

```python
image4 = cv2.dilate(image3,cv2.getStructuringElement(cv2.MORPH_CROSS,(5,5)),iterations=3)
plt.imshow(image4,cmap='gray')
cv2.imwrite('out14.jpg',image4)
```

```python
image5=cv2.morphologyEx(image4, cv2.MORPH_OPEN, cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (8, 8)))
image5=cv2.morphologyEx(image5, cv2.MORPH_CLOSE, cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (8, 8)))
plt.imshow(image5,cmap='gray')
cv2.imwrite('out15.jpg',image5)
```
